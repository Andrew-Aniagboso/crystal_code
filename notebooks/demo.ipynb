```python
from crystal_code.crystal import Crystal
import numpy as np

L = np.diag([4.0, 4.0, 4.0])
X = np.array([[0.0, 0.5],
              [0.0, 0.5],
              [0.0, 0.5]])
A = ['Na', 'Cl']

M = Crystal(X, A, L)
M
```

```python
print(M.cartesian_positions())
```

```python
M.save_cif("NaCl.cif")
```

```python
import matplotlib.pyplot as plt

X_cart = M.cartesian_positions()

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Draw lattice cube
ax.plot([0, L[0,0]], [0, L[1,0]], [0, L[2,0]], 'r-')
ax.plot([0, L[0,1]], [0, L[1,1]], [0, L[2,1]], 'g-')
ax.plot([0, L[0,2]], [0, L[1,2]], [0, L[2,2]], 'b-')

# Plot atoms
ax.scatter(X_cart[0], X_cart[1], X_cart[2], color='purple', s=200)
for i, label in enumerate(A):
    ax.text(X_cart[0,i], X_cart[1,i], X_cart[2,i], label, color='black')

ax.set_xlabel('x (Å)')
ax.set_ylabel('y (Å)')
ax.set_zlabel('z (Å)')
plt.show()
```
